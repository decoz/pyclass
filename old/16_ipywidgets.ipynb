#16 ipywidgets

#### UI (User Interface - 사용자 인터페이스) 란? 

초창기 프로그램이 정해진 일을 위주로 하는 '배치작업' 위주의 프로그램이었다고 하면 현대 프로그램은 사용자와의 상호작용하에 작동을 하는 경우가 많습니다. 상호작용이란 사용자의 키보드 입력, 바우스 입력, 메뉴선택등이 프로그램의 수행에 영향을 주는 것을 의미합니다. 이런 상호작용을 하게 해주는 프로그램의 부분을   **사용자 인터페이스(UI)** 라고 합니다. 
사실 파이썬은 대화영 프로그램이며 그 태생부터 자신이 원하는 작업을 스스로 수정해서 작동하는 방식으로 애용되었기 때문에 이런 UI 에 대한 필요성이 적었으며 그러기에 UI쪽은 지원이 강하지 안은 편입니다. 

하지만 그럼에도 매번 코드를 수정하는 것보다는 간단한 UI가 있는 것이 편한 경우가 있기에 간단한 UI 지원 라이브러리가 여럿 있습니다. 그 중에서 현재 여러분들이 사용하는 colab 환경에 맞춰 간단히 사용이 가능한 ipywidgets  라이브러리를 사용하여 간단한 UI를 작성하고 상호작용 프로그램의 경험을 해보도록 하겠습니다. 

먼저 ipywidgets 라이브러리와 이를 화면에 표시하기 위한 display 를 호출하겠습니다. 

````python
import ipywidgets as wgt
from IPython.display import display
````

## 16.1 위젯

위젯이란 사용자와의 상호작용을 받기 위한 하나의 객체정도로 이해하시면 됩니다. ipywidgets에는  버튼, 스크롤바, 슬라이더, 입력창 등등 다양한 위젯이 존재합니다. 
간단히 다음 코드를 보도록 하겠습니다. 



````python
btn = wgt.Button(description = "click!")
display(btn)
````

위의 코드는 btn 이라는 버튼 위젯을 생성하고 이를 화면에 표시합니다.  이렇게 입력의 단위가 되는 요소를 위젯이라고 합니다. 우리가 사용하는 프로그램들은 일반적으로 복수의 다양한 위젯들을 통해 상호작용합니다.




## 16.2 이벤트와 콜백함수 

하지만 위의 버튼을 클릭해도 아무 일도 일어나지 안습니다.  이번에는 아래 코드를 실행한 후에 위에 만든 버튼을 클릭해보시길 바랍니다. 

````python
def btn_click(v):
  print("clicked")
btn.on_click( btn_click )
````

이제 위의 화면에 버튼을 클릭시 clicked 라는 문자가 출력되는 것을 보실 수 있을겁니다. 
```
btn.on_click( btn_click )
```
은 버튼클릭시 인자로 받은 함수를 실행하라는 뜻입니다.  btn_click 은 물론 직접 호출도 가능하지만 이 경우는 버튼이 클릭시 작동할 수 있도록 이 함수를 버튼에게 전달하는 인자로 사용되었습니다.  파이썬에서는 이렇게 함수도 하나의 변수처럼 전달이 가능합니다. 특히 위와 같이 프로그램 외부에서 실행할 수 있도록 이를 인자로 넘겨주는 함수를 **콜백함수** 라고 표현합니다. 

그리고 위의 버튼클릭 같이 프로그램 외적인 특정 상황이 발생하는 경우를 **이벤트** 라고 표현합니다. 그리고 이런 이벤트상황에서 작동하는 콜백함수를 **이벤트 핸들러** 라고도 합니다. 

이제 UI 를 작성하는 방법을 간단히 정리해보도록 하겠습니다. 

1. UI 위젯 생성 
2. UI 위젯 출력
3. 이벤트 핸들러 함수 작성
4. 위젯에 이벤트 핸들러를 등록 

## 16.3 여러가지 위젯의 사용법 

ipywidgets 에서 제공하는 위젯은 매우 광범위하며 종류도 엄청 많습니다. 하지만 여기서는 대표적으로 많이 사용되는 몇몇 위젯들을 알아보고 간단한 사용법을 익혀보도록 하겠습니다. 

### 16.3.1 버튼 

위의 예제에서도 나왔던 버튼입니다.  버튼은 비교적 단순한 객체로 그 이벤트와 속성도 단순합니다.  다음은 버튼을 생성하고 그 안의 텍스트를 변경합니다. 

````python
btn2 = wgt.Button(description = "click!")
display(btn2)
btn2.description = 'abc'
````

이번에는 버튼을 눌렀을 경우 텍스트가 바뀌도록 해보겠습니다. 


````python
btn2 = wgt.Button(description = "click!")
display(btn2)
def btn2click(v):
  btn2.description = 'clicked'

btn2.on_click(btn2click)
````

참고로 다음 코드는 버튼을 사라지도록 합니다. 


````python
btn3 = wgt.Button(description = "hide")
display(btn3)
def btn3click(v):
  btn3.layout.display = 'none'

btn3.on_click(btn3click)
````

이제는 클릭하는 횟수만큼 버튼에 찍히는 버튼을 만들어보겠습니다. 

#### 함수와 글로벌 변수

함수의 내부에서 선언된 변수는 함수의 실행이 끝날때마다 파기됩니다. 그러므로 함수 내부의 변수로는 클릿 횟수를 저장할 수 없습니다.  이 경우는 함수 밖의 변수를 사용해야 하는데 외부 변수를 사용하기 위해서는 다음과 같은 문장이 필요합니다. 

```
global 변수명
```

그러면 이를 이용해 어떻게 카운트를 표시할 수 있는지 코드를 보겠습니다. 

````python
btn_c = wgt.Button(description = "0")
display(btn_c)
cnt = 0
def btnc_click(v):
  global cnt
  cnt += 1
  v.description = str(cnt)

btn_c.on_click(btnc_click)

````

### 16.3.2 레이블

위젯은 입력용만 있는것이 아닙니다. 텍스트를 보여줄 수 있는 출력도 있습니다. 이는 print같이 계속 텍스트를 누적해서 출력하지 안고 특정 텍스트를 보여주거나 교체하는데 사용됩니다. 



````python
label = wgt.Label(value = 'my label') 
label.layout.font_weight = 'bold'
display(label)
````

레이블 위젯은 텍스트 표시만을 위해 존재하는 위젯이며 .value 속성값을 변경해 표시된는 값을 바꿀 수 있습니다. 

````python
label = wgt.Label(value = '안누름') 
btn = wgt.Button(description = 'click')
def btnclick(v):
  label.value = '누름'

btn.on_click(btnclick)
display(label,btn)
````

#### 팁: 여러 버튼을 동시에 처리하기

### 16.3.3 슬라이더

이번에는 일정 범위의 정수를 입력할 수 있는 슬라이더입니다. 

````python
sld = wgt.IntSlider()
display(sld)
````

위와 같이 기본은 0 부터 100까지의 정수를 조절할 수 있습니다. 하지만 min, max 옵션으로 범위를 바꿀 수도 있습니다. 

````python
sld = wgt.IntSlider(min = -10, max = 10)
display(sld)
````

다음과 같이 기본 값을 설정할 수도 있습니다. 

````python
sld = wgt.IntSlider(min = -10, max = 10, value = 7)
display(sld)
````

이제 이벤트와 핸들러를 설정하는 법을 알아보겠습니다. 우선 아래의 코드를 먼저 실행해보죠.

````python
sld = wgt.IntSlider()
display(sld)

def onslide(v):
  print(v)
sld.observe(onslide)

````

한번 클릭으로도 상당히 많은 출력이 이뤄지는 것을 알 수 있습니다. 위에서 보듯이 observe 는 여러 이벤트에 대해 모두 발동하게 됩니다. 이때 값의 변화를 간단하게 얻어내기 위해서는 

> 'name' : 'value' 

형태의 v 값만 필요합니다. 물론 if 문으로 다음과 같이  처리할 수도 있지만 
```python
if v.name == 'value' :
  ...
```
observe 에서 더 간단하게 설정할 수 있습니다. 

````python
sld = wgt.IntSlider()
display(sld)

def onslide(v):
  print(v)
sld.observe(onslide, names = 'value')
````

이제 이벤트 인자 전체가 아니라 값만 출력하도록 하겠습니다. 




````python
sld = wgt.IntSlider()
display(sld)

def onslide(v):
  print(v.new)
sld.observe(onslide, names = 'value')
````

label 과 같이 사용하면 다음과 같이 사용이 가능하다. 

````python
sld = wgt.IntSlider()
label = wgt.Label(value = '0')
display(label, sld)

def onslide(v):
  label.value = str(sld.value)
sld.observe(onslide, names = 'value')
````

이번에는 슬라이더를 이용해 흑백 이미지의 밝기를 조정하는 프로그램을 보자

````python
from PIL import Image
import numpy as np 

arr = np.zeros((100,100,3)).astype('uint8')

sld = wgt.IntSlider(max = 255) 
display(sld)
def sliding(v):
  global arr
  arr[:,:,:] = v.new
  display(Image.fromarray(arr))

sld.observe(sliding, names = 'value') 
````

### 16.3.5 셀렉트

이번엔 여러 항목중에 하나를 선택할 수 있는 선택박스를 보겠습니다.  선택박스 역시 사용법은 크게 어렵지 안습니다만 선택항목을 잘 관리하면 매우 응용범위가 넓습니다. 



````python
select = wgt.Select(options = [1,2,3])
display(select)
````

이제 항목을 선택할때 발생하는 이벤트를 보도록 하죠. 아래 코드를 클릭하고 항목을 바꿔보세요. 

````python
def onselect(v):
  print(v)
select.observe(onselect, names = 'value')
````

마찬가지로 observe( names = 'value') 로 쉽게 값을 사용할 수 있습니다.  이제 특정 항목을 선택시 원하는 명령을 실행하게 하는 내용은 쉽게 사용하실 수 있을 겁니다. 

대신 특별히 옵션 리스트 자체를 변하게 하는 프로그램을 구경해보죠.




````python
select = wgt.Select(options = ['a','b','c'])
txt = wgt.Text()
btn = wgt.Button(description = 'add')

def addoption(v):
  if txt.value != '':
    select.options = list(select.options) + [txt.value]

btn.on_click(addoption)
display(select,txt,btn)


````

위의 코드는 아래쪽 텍스트입력창의 내용을 add 버튼으로 선택 박스에 추가하게 해줍니다.  아마 대부분은 크게 어렵지 안으시겠지만 이해가 안가는 부분도 있을겁니다. 
```python
select.options = list(select.options) + [txt.value]
```
select.options 는 배열처럼 보이지만 사실은 tuple 이라는 조금 다른 자료형을 사용합니다. 그래서 바로 아이템을 추가하기 어렵습니다. 그래서 int() 나 str() 
처럼 list() 를 이용해 이것을 배열형태로 변경이 가능합니다. ( tuple <- list 의 입력은 문제가 없습니다. ) 그래서 options 를 배열로 변환해 추가한 후에 다시 새로 교체해주는 것입니다.

이제 위의 선택창에서 디렉토리를 클릭시 해당 디렉토리로 이동하도록 해보겠습니다. 이를 위해서는 먼저 경로의 원리를 아셔야 합니다. 

현재 폴더는 '/content' 라는 경로이며 만일 sample_data 라는 폴더로 이동시에는 경로가  '/content/sample_data' 가 됩니다. 그러므로 다음 과 같은 같이 새 경로를 지정할 수 있습니다. 

```python
cpath += '/' + new_dir 
```

이제 select 에 이벤트 핸들러를 설정해서 위에 sample_data 디렉토리를 클릭시 그 안으로 이동하도록 해보겠습니다.  아래의 코드를 실행 후에 select 에 항목에서 sample_data 를 클릭해보시면 이제 그 안의 항목으로 변경되는 것을 보실 수 있습니다. 


````python
def change_dir( v ):  
  global cpath
  if os.path.isdir( v.new ) : # 클릭한 항목이 폴더면 
    cpath = cpath + "/" + v.new # 패스를 추가
    flist = os.listdir( cpath ) 
    fsel.options = flist

fsel.observe(change_dir, names = 'value')
````

이 부분을

이렇게 해주면 선택 항목에 .. 항목이 생기고 이를 클릭시 상위 폴더로 가게 됩니다.

````python
cpath = os.getcwd()
flist = os.listdir(cpath)
fsel = wgt.Select( options  = ['..'] + flist, layout = {'height':'400px'} )
display(fsel)

def change_dir( v ):  
  global cpath
  if os.path.isdir( cpath +"/" + v.new ) : # 클릭한 항목이 폴더면 
    cpath = cpath + "/" + v.new # 패스를 추가
    fsel.unobserve(change_dir, names = 'value')    
    flist = os.listdir( cpath ) 
    fsel.options = ['..'] + flist
    fsel.value = None
    fsel.observe(change_dir, names = 'value')

fsel.observe(change_dir, names = 'value')
````

#### <font color = 'red'> 연습 16.6.2 : 폴더에는 [] 붙여서 표기하기
위의 탐색기는 폴더와 파일이 구분이 안되는 단점이 있습니다. 폴더의 경우 다음 [] 를 붙여서 표기가 되도록 위의 탐색기를 수정하세요
</font>

image.png



````python
# 연습문제 16.5.1 의 코드를 작성새보세요

def transdir(flist, path):
  for i in range( len(flist) ):
    if os.path.isdir(path + "/" + flist[i]) :
      flist[i] = '['+flist[i]+']'
  
cpath = os.getcwd()
flist = os.listdir(cpath)
transdir(flist, cpath)
fsel = wgt.Select( options  = ['[..]'] + flist, layout = {'height':'400px'} )
display(fsel)


def change_dir( v ):  
  global cpath

  if v.new[0] == '[' and v.new[-1] == ']' : 
    cpath = cpath + "/" + v.new[1:-1]   

    fsel.unobserve(change_dir, names = 'value')    
    flist = os.listdir( cpath ) 
    transdir(flist, cpath)

    fsel.options = ['[..]'] + flist
    fsel.value = None
    fsel.observe(change_dir, names = 'value')

fsel.observe(change_dir, names = 'value')



````

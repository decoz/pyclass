## 10.1 함수

함수는 프로그램의 일부를  변수처럼 이름을 붙여서 정의하고 필요할때 이를 몇번이고 사용할 수 있게 한 것입니다.  




### 10.1.1 함수 정의방법

일반적인 함수는 다음과 같은 형식으로 정의됩니다.

```python
def 함수명(인자):
  코드
  ...
  return 반환값
```

- 함수명 : 함수의 이름으로 일반적인 변수 작성 규칙과 유사
- 인자 : 함수를 호출할 때 입력되는 변수값의 종류
- 반환값 : 함수가 실행을 마쳤을때 이 함수를 호출한 위치에 남기는 값

아직은 이해가 잘 안가실 수 있는데 먼저 간단한 예제를 보도록 하겠습니다.


````python
def hello() :
  print("hello")
````

위에서 정의된 함수 hello 는 인자도 반환값도 없는 간단한 함수로서 실행내용도 화면에 'hello' 를 출력하는 함수입니다. 하지만 중요한 것은 위의 코드를 실행해도 화면에는 아무것도 출력되지 안습니다.  왜냐하면 위의 코드는 어디까지나 함수를 '**정의**' 했을 뿐이며 이것이 실행되려면 정의된 함수의 이름을 이용해야 합니다. 이렇게 정의된 함수를 실행하는 것을 '**호출**' 이라 합니다.  

### 10.1.2 함수의 호출


함수를 호출할때는
```python
함수명(인자값)
```
로 () 를 붙여서 그 안에 필요 인자값을 넣어서 실행을 시킵니다.

다만 현재 위의 hello처럼 인자가 없는 경우는 () 로 비워둔 형식으로 호출하면 됩니다. 한번 정의한 함수는 몇번이든 호출이 가능합니다.


````python
hello()
hello()
````

### 10.1.3 함수의 인자

인자(parameter)는 각 함수가 실행하기 위해 입력해야 하는 값을 의미합니다. 이 인자를 사용해서 함수가 더 다양한 기능을 할 수 있도록 할 수 있습니다.

그러면 이번에는 인자를 사용하여 위의 hello 를 좀더 쓸모있게 만들어보겠습니다.



````python
def hello(who) :
  print("hello", who)

hello('mom')
hello('dad')
````

이렇게 이젠 함수를 호출할때 그 안에 사람이름을 넣어주면 그 사람에 대한 인삿말이 출력됩니다.

**인자의 기본값**

그런데 만일 10번중에 8번가량은 그냥 hello만 출력하고 나머지 2번가량만 뒤에 이름을 붙여서 출력하는 프로그램을 작성한다면 .. 그러면 20%의 'hello 이름' 이 필요한 경우를 위해서 나머지 80%를 매번 hello('')로 호출하는 것은 괴로운 일일겁니다.  

이 경우 인자의 기본값을 설정하면 hello() 로 호출시 who에 자동으로 ' ' 값이 입력되게 할 수 있습니다.



````python
def hello(who = '') :
  print("hello", who)

hello('mom')
hello()
````

또한 인자는 여러개를 사용할 수도 있습니다. 아래의 함수는 두 숫자를 입력받아서 그 중에 큰 숫자를 출력해줍니다.


````python
def max_print(a,b):
  if a > b :
    print(a)
  else:
    print(b)

max_print(4,5)
````

### 10.1.4 함수의 반환값

함수를 호출하는 측에서 함수에 전달하는 자료가 '**인자값(parameter value)**' 이라면 함수가 실행을 마치고 호출하는 측에 전달하는 값을 '**반환값(return value)**' 이라고 합니다. 이는 함수 내부에서 다음과 같이 사용합니다.

```python
return 값
```

주의할건 return 이 수행되면 값을 반환하고 함수는 작동을 멈춥니다. (이를 이용해 그냥 return 만 써서 함수를 멈추는 방법도 사용됩니다. )


````python
def one():
  return 1

a = one()
print(a)
````

이러한 리턴값을 갖는 함수는 변수처럼 값으로 사용도 가능합니다.
```python
a = 1
a = a + one()
```

이런 인자값과 리턴값을 잘 사용하면 하나의 기능이라도 다양한 경우에 사용합니다. 예를 들어 max_print 는 더 큰 값을 출력하는 일 밖에 못하지만 대신 다음과 같이 maxv 라는 함수를 만들어서 사용하면 출력 뿐만 아니라 더 큰 값을 다른 변수에 저장하거나 둘 이상의 숫자중에 더 큰 숫자를 얻을 수도 있습니다.

````python
def maxv(a,b):
  if a > b:
    return a
  else :
    return b

print(maxv(5,4))
````

예를 들어 아래처럼 사용하면 3이상의 숫자중에 가장 큰 값도 구할 수 있습니다.

````python
print( maxv( maxv(7,8), 9 ) )
````

사실 더 큰 값을 구하는 기능은 매우 많이 사용하기 때문에 파이썬에는 이미 max 라는 함수가 기본으로 제공되며 이 max 는 임의의 숫자의 인자도 사용가능합니다. 또한 배열도 사용가능합니다. ( 여러분도 이런 함수의 제작이 가능하지만 이런 함수인자의 고급 사용법은 아직 여러분께는 필요치 안으니 여기서는 생략합니다. )

````python
print('basic max:', max(3,4))
print('multiple max:', max(5,7,2,4))

arr = [4,7,2,8]
print('array:', max(arr))
````

### 10.1.5 함수의 독립성과 global

함수는 호출이 됨과 동시에 함수에 사용하는 변수영역이 새로 생성되며 함수는 일종의 독립공간으로 간주됩니다. 그래서 원칙적으로는 함수는 일반 영역에 있는 변수를 사용할 수 없으며 함수가 작동하는데 필요한 외부값은 인자로 넣어줘야 합니다.

다음 프로그램을 한번 실행해보죠

````python
a = 3
def f():
  a = 1

f()
print(a)
````

함수 f가 실행되었지만 a의 값은 변하지 안은 것을 알 수 있습니다. 이는 함수 내부의 a 와 밖의 a는 서로 관계가 없는 독립된 존재이며 함수 안의 a 는 f가 실행을 마친 순간 파기 됩니다.

하지만 어떤 기능이 외부에 있는 자료를 너무 많이 사용할 경우 이를 호출할때마다 인자로 다 넣어주는 것은 너무 번거로운 일입니다. 이 경우 globaㅣ 로 외부 변수의 사용을 요청할 수 있습니다.


````python
a = 3
def f():
  global a
  a = 1

f()
print(a)
````

### 10.1.6 재귀함수

함수는 내부에서 다른 함수를 호출할 수도 있지만 자기 자신을 호출할 수도 있습니다.이를 재귀 호출이라고 합니다.

```python
def f(): # 절대로 실행해서는 안되는 함수
  f()
```

위의 함수는 무한히 자기자신을 호출하기 때문에 끝나지 안는 무한루프현상을 일으킵니다. 하지만 이를 잘 제어하면 반복문 이상의 일을 할 수 있습니다.




````python
def count(n):
  print(n)
  if n > 0 :
    count(n-1)

count(10)

````

위의 count는 자기 자신을 호출하지만 인자인 n의 값을 하나씩 줄여서 호출하며 n이 0 이 되면 자기 복제를 멈추기 때문에 무한루프를 방지할 수 있습니다.  

효율적이라고 하기는 힘들지만 다음의 재귀함수는 피보나치 수열을 상당히 아름답게 작성 가능합니다.



````python
def fib(n):
  if n <= 2 :
    return 1
  else:
    return fib(n-1) + fib(n-2)

for i in range(1,10):
  print(fib(i))
````

재귀 함수의 가치는 다음과 같은 중첩구조에서 그 진가를 발휘합니다.

````python
arr = [1,2,[3,4,[5,10,15]],[2,4,[7,8,9]]]
def sumarr(arr):
  if type(arr) != list:
    return arr
  sum = 0
  for a in arr:
    sum += sumarr(a)
  return sum
print( sumarr(arr) )
````

````python
arr = [1,2,[3,4,[5,10,15]],[2,4,[7,8,9]]]

def printarr(d, arr):
    if type(arr) != list:
      print('  ' * d, arr)
    else:
      for a in arr:
        printarr(d+1, a)
printarr(0,arr)
````

*주의:
재귀 호출은 이 경우 함수가 끝나지 안은 상태에서 자신이 또 호출되므로 변수공간이 배열처럼 중첩되서 생깁니다. 이 때문에 잘 못 사용하면 무한히 자기자신을 생성해서 무한루프 현상이 발생합니다. 그냥 헛도는 무한 반복과는 달리 재귀함수의 무한호출은 계속해서 변수영역을 무한히 생성하므로 시스템의 메모리를 급격하게 점유해가며 시스템 전체를 다운시킬 수 있기때문에 조심스럽게 작성해야 하며 무한에 빠지면 최대한 빨리 강제정지를 해줘야 합니다.*

## 10.2 클래스

프로그램의 초창기엔 프로그램을 기능별로 함수같은 서브루틴으로 제작하여 프로그램을 나누어 작성했습니다. 그런데 프로그램이 점점 커지고 복잡해지면서 각 기능별로 주고 받아야 하는 자료의 양도 커지고 복잡해 졌습니다.

어떤 프로그램을 기능별로 나눌 경우 기능에 관련된 자료들과 이 자료들을 처리하는 함수들을 모아서 하나의 모음으로 처리하는게 더 편리하다는 것을 알게됩니다. 그래서 자료와 함수를 모아서 '객체' 라는 개념으로 정의하고 이 객체단위로 프로그램을 구성함으로서 더 직관적이면서도 재사용이 편해지게 됩니다.

객체는 다음 두 단계로 생성됩니다.

- 객체의 유형인 '클래스' 를 정의
- 클래스에 정의된 객체를 실제로 생성

먼저 클래스의 선언은 다음과 같은 형식으로 이뤄집니다.

### 10.2.1 클래스와 객체

```python
class 클래스():
  
  변수명 = 초기값
  :
  
  def 함수명(self, 인자들) :
    함수 정의   
  :
```

그러면 실제 간단한 클래스 하나를 생성해보겠습니다.




````python
class myclass():
  a = 1
  b = 2

my = myclass()
print(my.a, my.b)
````

myclass 는 클래스 .. 즉 객체의 유형이 되고.. my는 이 유형으로 생성된 자료입니다.  이렇게 생성된 객체는 myclass 안에 정의된 a,b 를 변수를 사용할 수 있는데 이런 객체 내분의 변수를 '멤버변수' 라고 합니다.



### 10.2.2 메소드

객체에 포함된 변수를 '멤버변수' 라 부르듯이 객체의 함수는 '멤버함수' 또는 '메소드' 라고 부릅니다. 메소드의 정의는 일반 함수와 비슷하지만 다음과 같은 특징이 있습니다.

- 함수의 첫인자는 반드시 self 가 되어야 합니다.
- 함수는 객체의 멤버변수를 self.변수명 으로 사용가능합니다.

그러면 위에 myclass 에 출력을 담당하는 print라는 메소드를 추가해보도록 하겠습니다.


````python
class myclass():
  a = 1
  b = 2
  def print(self):
    print(self.a, self.b)

my = myclass()
my.print()
````

한번 클래스를 정의하면 이 클래스로 여러개의 객체(인스턴스)를 만들 수도 있습니다.

````python
my1 = myclass()
my1.a, my1.b = 3,4
my2 = myclass()
my2.a, my2.b = 5,6

my1.print()
my2.print()
````

### 10.2.3 생성자

위의 예제에서 본 객체는 단순하지만 만일 상당히 복잡한 기능의 객체를 정의할 경우 객체의 초기 멤버변수 값을 입력하거나 또는 초기화 과정이 필요할 수가 있습니다. 이 경우 생성시에 작동하는 함수를 메소드처럼 정의할 수 있는데 이를 '생성자' 라고 합니다. 생성자는 다음과 같이 메소드를 정의합니다.

```python
class 클래스명():
  def __init__(self, 추가 인자들):
    생성시 작동내용
```

그러면 한번 예를 보도록 하겠습니다. 위의 myclass 예제에서 myclass 의 a,b 값을 매번 객체를 생성한 후에 따로 넣어주는게 귀찮다면  생성자를 이용해 다음과 같이 사용할 수 있습니다.


````python
class myclass():
  a = 1
  b = 2

  def __init__(self, a, b):
    self.a = a
    self.b = b


  def print(self):
    print(self.a, self.b)

my = myclass(10,20)
my.print()
````

__init__ 는 객체가 생성되는 시점인 myclass(10,20) 이 호출될때 실행되며 만일 생성자를 위의 예제처럼 정의했다면 myclass 는 반드시 2개의 값을 입력해줘야 객체가 생성됩니다.

### 10.2.4 상속

만일 우리가 학교시스템을 만들기 위해 학부생과 대학원생을 객체로 정의한다고 생각해봅시다. 학부생과 대학원생은 상당히 많은 부분에 공통점을 가지고 있지만 대학원생은 약간의 추가적인 정보가 더 필요하다고 해보죠.
```
학부생 :
  학번 , 이름, 이수과목(배열)

대학원생:
  학번, 이름, 이수과목(배열), 연구실, 지도교수
```

이 경우 두가지 방법을 생각할 수 있습니다.

- 학부생과 대학원생의 객체를 따로 정의
- 모든 학생을 대학원생 전용정보까지 포함하게 만듬

첫번째 방법은 모든 관련 함수와 메소드를 따로 작성해야 하므로 유사한 코드를 각각 반복해야 합니다.  만일 수정해야 할 내용이 생기면 역시 각각 수정을 해줘야하죠
두번째 방법은 공통기능은 일원화할 수 있지만 대신 소수의 대학원생을 위해서 수많은 학부생들이 의미없는 변수를 포함하게 됩니다.

대학원생을 정의할때 학부생의 객체의 공통부분을 그냥 포함해 쓰면 어떨까요? 그러면 학부생 기준으로 처리되는 모든 함수는 그대로 사용이 가능하면서 대학원 전용 기능에만 대학원용 기능을 사용할 수 있다면 상당히 효율적일 겁니다.

이를 객체에서는 '상속' 이라는 기능을 통해 구현합니다. 상속은 자식 클래스가 부모 클래스의 변수와 함수를 다 사용할 수 있게 해줍니다.

```python
class 부모():
  ...

class 자식(부모)
  ...
```

이제 이를 이용해 위의 학생용 객체를 정의해보도록 하겠습니다.










````python
class stu(): # 학부생(부모) 클래스

  lectures = []

  def __init__(self, no, name):
    self.no = no
    self.name = name

  def info(self):
    s = 'name:' + self.name + ' no:' +  self.no
    return s


class grad(stu): # 대학원생(자식) 클래스

  def __init__(self, no, name, lab, professor):
    super().__init__(no,name)
    self.lab = lab
    self.professor = professor


kay = stu('2377018', '김아영')
psy = grad('2388102', '박서윤', '인지과학연구실', '김치민')

print(kay.info())
print(psy.info())
````

클래스를 잘 정의해 자료와 기능들을 잘 분리해 구성을 하면 거대한 프로그램을 작성하고 관리하는데 드는 시간과 오류의 가능성을 매우 줄일 수 있습니다.  이러한 식으로 프로그램을 개발하는 것을 '객체지향' 이라고 하며 시중에는 수많은 객체지향 설계기법에 대한 서적이 나와있습니다.

하지만 객체지향은 복잡하고 거대한 프로그램을 작성해본 경험이 없으면 그 필요성과 활용방법을 이해하기 어렵기 때문에 배움의 단계에 있는 여러분들은 굳이 객체지향을 크게 염두에 두지 안으셔도 됩니다. 하지만 앞으로 여러 라이브러리를 사용할때 대부분 이러한 클래스로 기능이 제공되므로 숙련된 프로그래머들이 어떻게 객체를 만들어서 여러사람들이 편하게 기능을 사용하게 했는지를 체험하시다 보면 조금씩 그 의미를 알아가실 수 있을겁니다.

## 5 배열 ( List ) 

배열은 복수의 값을 저장할 수 있는 자료형을 의미합니다. 현대의 컴퓨터가 처리하는 자료는 대부분 대용량이므로 이를 하나하나 변수에 저장할 수 없기 때문에 배열 ( 이것도 엄밀히는 변수의 일종입니다. ) 에 저장하게 됩니다. 

다음은 간단한 배열의 예입니다. 

```
a = [1,2,3]
```

위에서 a는 변수지만 그 안에 1,2,3 이라는 3가지 값을 담고 있습니다. 하나의 배열에 들어있는 여러 값을 [ ] 형태로 표기하곤 합니다. 이 배열의 가각 개별 값을 억세스 할 때는 다음과 같이 합니다. 

```
print( a[0], a[1] ) 
```
위에서 a[0] 는 a 의 값들 중에 첫번째 값을 의미합니다.  이렇게 여러개의 값을 지닌 배열에서 특정 정보를 얻어낼 때는 **배열이름[번호]** 형태로 표기합니다. 주의할 것은 항상 0번부터 시작한다는 것 입니다. 즉 첫번째 값은 a[1] 이 아니라 a[0] 이 되며 마찬가지로 세번째 값은 a[2] 가 됩니다. 

배열은 그 값을 추가하거나 뺄 수도 있습니다. 

```
a = a + [4,5] 
```
위의 예에서 a + [4,5] 는 배열 + 배열로서 두 배열을 합치는 것으로 처리됩니다.  즉 a = [1,2,3,4,5] 가 되게 되는 것이죠 
```
del( a[0] ) 
```
위의 문장은 배열 a에서 첫번째 값을 뺀다는 의미입니다.  그러면 한번 아래의 코드를 돌려서 직접 확인해보세요











````
a = [1,2,3]
a = a + [4,5]
print(a)

del(a[0])
print(a)
````

a[0] , a[1] 처럼 배열에서 몇번째인지를 의미하는 에서 [0] 이나 [1] 은 인덱스 값이라고 하는데 이때 이 값은 연산이 가능합니다.  a[1+1] 같은 단순 계산도 가능하지만 무엇보다 이때 변수값을 사용이 가능합니다.

배열의 값들은 숫자가 될수도 있지만 다시 다른 배열을 가질 수도 있습니다. 이를 다중 배열이라고 합니다.  다음 예를 보죠 

```
a = [ [1,2,3], [4,5,6] ]
```
위의 예제에서 a는 두개의 값을 갖은 크기 2의 배열입니다. 그런데 두개의 값은 일반 숫자가 아닌 다시 다른 배열인 것이죠. 그래서 a[0] 은 [1,2,3]  , a[1] 은 [4,5,6] 이 됩니다. 이렇게 배열속에 다시 배열을 갖는 구조를 다중 배열이라고 합니다. 

이러한 다중배열속의 실제 숫자 값을 꺼내는 것은 인덱스도 하나 이상이 필요합니다. a[0] 인 [1,2,3]의 첫번째 숫자를 꺼내려면 a[0][0] 라고 쓰면 됩니다. 

이러한 다중배열은 실제로 컴퓨터에서 2차원 데이터를 다루는데 많이 사용됩니다. 예를 들어 이미지같은 경우는 가로 x 세로 점들의 모음이므로 그 구조는 기본적으로 2차원입니다.  거기에 컬러 이미지의 경우는 가로 x 세로의 점 마다 다시 3가지 색상정보 (RGB)값을 갖고 있으므로 실제로는 3차원 구조죠.  이런 배열을 다룰때는 다중루프가 효과적입니다.  

예를 들어 위의 a 는 2 x 3 의 구조입니다만 각 내부의 크기가 같으란 보장은 없습니다.  이런 a값을 다 억세스하려면 다음과 같은 코드가 효과적입니다. 아래 코드의 패턴은 이런 다중 배열을 다루는 대표적인 구조이므로 직접 한번 실행해보시길 바랍니다. 


````
a = [[1,2,3], [4,5]]
i = 0
while i < len(a):
  j = 0 
  while j < len(a[i]): # a의 내부 배열의 길이
    print(a[i][j], end = ' ')
    j = j + 1
  print()
  i = i + 1
````

위의 "while j < len( a[i] ) " 는 실제 a 배열의 인자인 [1,2,3]  과 [4,5] 의 크기를 구해주기 때문에 길이가 다른 배열도 처리를 가능하게 합니다. 이렇게 다중배열은 다중루프로 처리하는게 효과적이므로 이런 다중루프를 잘 다루는가가 컴퓨터의 자료처리를 자율롭게 할 수 있는 프로그램 알고리즘의 기초가 됩니다.

## 6 for 문

while 은 루프의 기초지만 배열값을 억세스 하기 위해 여러줄의 코드가 필요하다는 단점이 있습니다. 이렇게 배열값들을 억세스할때 while 보다 더 간단히 쓰기 위해 만들어진 코드가 for 문입니다. 

만일 어떤 배열 a 가 있다고 가정하면 
```
a = [1,2,3,4,5]
```

다음 문장은 이 배열의 값을 하나하나 출력하게 됩니다. 
```
for v in a :
    print(v)   
```
위의 문장에서 for 문은 a 의 값을 하나씩 v 에 넣어주며 반복을 수행합니다. 이는 실제로 다음 while 문과 같은 역할을 합니다. 

```
i = 0
while i < len(a):
  v = a[i]  
  print(v)
  i = i + 1
```

이와 같이 for 문은 배열을 더 쉽게 다룰 수 있도록 while문을 더 간결하게 만든 개념이라고 보시면 됩니다.  하지만 for 문이 반드시 배열에만 특화된 것은 아닙니다. 

```
for v in range(0,5):
  print(v) 
```

위의 예에서 range 는 0,1,2,3,4 라는 일종의 가상배열을 만들어줍니다. 그래서 for 문은 0~4 를 찍어줍니다.  이는 이와 대응하는 while 문보다 더 간결합니다. 

```
i = 0
while i < 5 :
  print(i)
  i = i + 1  
```

이를 통해 for 문은 배열을 억세스 하거나 일정 숫자를 증가 감소시킬때 유용합니다. range는 단순히 1씩만 증가하는 것도 아닙니다.  

```
# 10부터 99까지 2씩 증가 
for v in range(10,100,2): 
  print(v) 


# 100부터 1 까지를 출력
for v in range(100,0,-1): 
  print(v) 

```

반복문을 사용하는 경우의 많은 경우는 배열을 억세스 하거나 고정 횟수를 일정 패턴으로 반복시키는 경우가 많기 때문에 실제 프로그래밍에서는 while 보다 간편한 for 를 더 많이 사용합니다.

 하지만 while 문은 조금 불편해도 더 많은 경우를 대응할 수 있습니다. 만일 누적 합산값이 100 이상인 가장 작은 짝수를 구한다고 해보죠
```
i,sum = 0,0
while sum < 100:
  sum = sum + i 
  i = i + 2
print(i)
```

위의 프로그램은 합산값이 sum 이 100 보다 작은 경우만 반복하므로 100 이상이 되는 순간 멈추고 이때 sum은 짝수 i값의 누적값이 되므로 우리가 원하는 답은 i가 됩니다. 

이렇게 어떤 상황이 발생할 때 까지 불특정 횟수를 반복하는 경우는 while 문이 꼭 필요합니다. 그리고 원리적으로도 while문이 더 컴퓨터의 작동방식에 가깝기 때문에 어디까지나 for 문은 while 문의 편의성을 높인 부분집합으로 보시는게 좋습니다. 

위의 이중 배열을 억세스하는 것도 이 경우는 매우 간결해집니다 

```
a = [ [1,2,3], [4,5] ] 
for arr in a :
  for v in arr:
    print(v) 
```
단 주의할 것은 v 는 어디까지나 a 안의 값을 복사한 값입니다. 

```
for arr in a :
  for v in arr:
    v = v + 1
```

는 실제로 배열값에 변화를 주지 못합니다. 이런 경우는 어쩔 수 없이 다음과 같이 작성하셔야 합니다. 

```
for i in range(0, len(a)):
  for j in range(0, len(a[i]) ):
    a[i][j] = a[i][j] + 1    
```

결국 배열 값 자체를 억세스하고 변화시키려면 인덱스 값을 얻어내야 한다는 것이죠.  그러면 이제 간단한 연습문제를 풀어보도록 합시다.
